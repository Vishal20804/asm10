```python
Ans:1
import scipy.stats as stats
import numpy as np

def variance_ratio_test(data1, data2):
    n1 = len(data1)
    n2 = len(data2)
    
    var1 = np.var(data1, ddof=1)  # unbiased estimate of variance for data1
    var2 = np.var(data2, ddof=1)  # unbiased estimate of variance for data2
    
    f_value = var1 / var2  # calculating the F-value
    
    df1 = n1 - 1  # degrees of freedom for data1
    df2 = n2 - 1  # degrees of freedom for data2
    
    p_value = 1 - stats.f.cdf(f_value, df1, df2)  # calculating the p-value
    
    return f_value, p_value
data1 = [1, 2, 3, 4, 5]
data2 = [2, 4, 6, 8, 10]

f_value, p_value = variance_ratio_test(data1, data2)
print("F-value:", f_value)
print("p-value:", p_value)
```

```python
Ans:2
import scipy.stats as stats

def critical_f_value(significance_level, dfn, dfd):
    alpha = significance_level / 2  # Two-tailed test, so divide significance level by 2
    critical_value = stats.f.ppf(1 - alpha, dfn, dfd)
    return critical_value
significance_level = 0.05
dfn = 5
dfd = 10

critical_value = critical_f_value(significance_level, dfn, dfd)
print("Critical F-value:", critical_value)
```

```python
Ans:3
import numpy as np
import scipy.stats as stats

def variance_equality_test(data1, data2):
    n1 = len(data1)
    n2 = len(data2)
    
    var1 = np.var(data1, ddof=1)  # unbiased estimate of variance for data1
    var2 = np.var(data2, ddof=1)  # unbiased estimate of variance for data2
    
    f_value = var1 / var2  # calculating the F-value
    
    df1 = n1 - 1  # degrees of freedom for data1
    df2 = n2 - 1  # degrees of freedom for data2
    
    p_value = 2 * (1 - stats.f.cdf(f_value, df1, df2))  # calculating the p-value for a two-tailed test
    
    return f_value, df1, df2, p_value

# Generate random samples from two normal distributions
mean1 = 0
variance1 = 1

mean2 = 0
variance2 = 1

sample_size = 20

data1 = np.random.normal(mean1, np.sqrt(variance1), sample_size)
data2 = np.random.normal(mean2, np.sqrt(variance2), sample_size)

# Perform variance equality test
f_value, df1, df2, p_value = variance_equality_test(data1, data2)

# Print the results
print("F-value:", f_value)
print("Degrees of freedom:", df1, ",", df2)
print("p-value:", p_value)

```
